```python
import itertools
import pandas as pd
import numpy as np

numList = list(range(1,7))
resultList1 = list(itertools.product(numList,repeat=5)) 

resultArray = np.array(resultList1)
sumOutcomes = np.sum(resultArray, axis=1)

df0 = pd.DataFrame({"Outcomes":resultList1,"Sum of Outcomes":sumOutcomes})
df0
```

```python
by_sumOutcomes = df0.groupby("Sum of Outcomes")
by_sumOutcomes.describe()
outcomeDF = by_sumOutcomes.count()
outcomes = outcomeDF.values[:,0]
probability = outcomes / np.sum(outcomes)
finalDF = pd.DataFrame({"sum":df0["Sum of Outcomes"].unique(), "Outcomes":outcomes, "Probability":probability})
finalDF
```
